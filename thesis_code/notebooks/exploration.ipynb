# ExG Preprocessing Exploration

This notebook loads sample leaf images, applies the Excess Green (ExG) mask, and visualizes the original image next to the masked result.

```python
import matplotlib.pyplot as plt
import cv2
from pathlib import Path
from src.preprocessing.exg import apply_exg

sample_dir = Path('data/raw')
images = []
for ext in ['*.jpg', '*.jpeg', '*.png', '*.bmp']:
    images.extend(sample_dir.rglob(ext))
images = images[:4]

fig, axes = plt.subplots(len(images), 2, figsize=(10, 5 * len(images)))
for idx, image_path in enumerate(images):
    image = cv2.cvtColor(cv2.imread(str(image_path)), cv2.COLOR_BGR2RGB)
    masked, mask = apply_exg(image)
    axes[idx, 0].imshow(image)
    axes[idx, 0].set_title('Original')
    axes[idx, 0].axis('off')
    axes[idx, 1].imshow(masked)
    axes[idx, 1].set_title('ExG masked')
    axes[idx, 1].axis('off')
plt.tight_layout()
plt.show()
```
